```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV file
df = pd.read_csv("/Users/tylukasiewicz/Downloads/NHL_Injury_Data - NHL Injury Database_Full Data_data.csv")

# Ensure 'Season' and 'Injury Type' are strings
df['Season'] = df['Season'].astype(str)
df['Injury Type'] = df['Injury Type'].astype(str)

# Filter for "Shoulder" injuries (or specific shoulder injury types if available in your data)
shoulder_injuries_df = df[df['Injury Type'].str.contains("Upper body", case=False, na=False)]

# Count the number of shoulder injuries per season
shoulder_injury_counts = shoulder_injuries_df['Season'].value_counts().sort_index()

# Convert the Series to a DataFrame for easier manipulation
shoulder_injury_counts_df = shoulder_injury_counts.reset_index()
shoulder_injury_counts_df.columns = ['Season', 'Injury_Count']

# Calculate the percentage increase in shoulder injuries between consecutive seasons
shoulder_injury_counts_df['Percentage_Increase'] = shoulder_injury_counts_df['Injury_Count'].pct_change() * 100

# Remove the first row which will have NaN for percentage increase as there's no prior season to compare
shoulder_injury_counts_df = shoulder_injury_counts_df.dropna()

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(shoulder_injury_counts_df['Season'], shoulder_injury_counts_df['Percentage_Increase'], marker='o', linestyle='-')

# Add labels and title
plt.xlabel('Season')
plt.ylabel('Percentage Increase in Upper body (%)')
plt.title('Percentage Increase in Upper body per Season')

# Rotate x-axis labels if they overlap
plt.xticks(rotation=45, ha='right')

# Add a horizontal line at 0 for easier interpretation
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)

# Display the plot
plt.grid(True)
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Load data
df = pd.read_csv("/Users/tylukasiewicz/Downloads/NHL_Injury_Data - NHL Injury Database_Full Data_data.csv")

# Ensure proper types
df['Season'] = df['Season'].astype(str)
df['Injury Type'] = df['Injury Type'].astype(str)

# Filter upper body injuries
upper_body_df = df[df['Injury Type'].str.contains("Upper body", case=False, na=False)]

# Count per season
injury_counts = upper_body_df.groupby("Season").size().reset_index(name="Injury Count")

# Split into before (2016/17–2019/20) and after (2020/21–2021/22+)
before = injury_counts[injury_counts['Season'].isin(["2016/17","2017/18","2018/19","2019/20"])]
after = injury_counts[injury_counts['Season'].isin(["2020/21","2021/22"])]

# Statistical test: Are they different?
t_stat, p_val = ttest_ind(before["Injury Count"], after["Injury Count"], equal_var=False)

print("Average injuries before rule:", before["Injury Count"].mean())
print("Average injuries after rule:", after["Injury Count"].mean())
print("T-test p-value:", p_val)

# Plot
plt.figure(figsize=(10,6))
plt.plot(injury_counts["Season"], injury_counts["Injury Count"], marker='o')
plt.axvline(x="2020/21", color='red', linestyle='--', label="Rule Change")
plt.title("Upper Body Injuries Per Season (NHL)")
plt.xlabel("Season")
plt.ylabel("Number of Injuries")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
